```python
# Import pandas library (used for data manipulation and analysis)
import pandas as pd

# Import NumPy library (used for numerical computations)
import numpy as np

# Import matplotlib's pyplot module (used for creating plots and graphs)
import matplotlib.pyplot as plt

# Import seaborn library (used for advanced data visualization)
import seaborn as sns
```

```python
# Load the built-in Titanic dataset from Seaborn library
# This dataset contains information about passengers such as age, sex, class, fare, and survival status
Data = sns.load_dataset('titanic')
```

```python
# Display the entire Titanic dataset in the output
# Shows all rows and columns (may be large depending on dataset size)
print(Data)
```

```python
# Display a concise summary of the dataset
# Shows column names, data types, non-null values, and memory usage
Data.info()
```

```python
# Generates statistical summary (mean, std, min, max, etc.) of numerical columns
Data.describe()
```

```python
# Display the data types of each column in the dataset
print(Data.dtypes)
```

```python
# Count the number of occurrences of each unique value in the 'pclass' column
Data["pclass"].value_counts()
```

```python
# Calculate the percentage of passengers who survived in 1st class (pclass = 1)
np.mean(Data['survived'][Data["pclass"] == 1]) * 100
```

```python
# Calculate and print the percentage of passengers who survived in 1st class
print(np.mean(Data['survived'][Data["pclass"] == 1]) * 100)
```

```python
# Calculate and print the percentage of passengers who survived in 2nd class
print(np.mean(Data['survived'][Data["pclass"] == 2]) * 100)
```

```python
# Calculate and print the percentage of passengers who survived in 3rd class
print(np.mean(Data['survived'][Data["pclass"] == 3]) * 100)
```

```python
# Calculate and print the average survival rate for male passengers
print("male =", np.mean(Data['survived'][Data['sex'] == 'male']))
```

```python
# Calculate and print the average survival rate for female passengers
print("female =", np.mean(Data['survived'][Data['sex'] == 'female']))
```

```python
# Print the total number of rows (records) in the dataset
print(len(Data))
```

```python
# Print the number of passengers who survived (survived = 1)
print(len(Data[Data['survived'] == 1]))
```

```python
# Create a count plot to show the number of passengers by gender (sex)
# and compare survival status (survived) using different colors
sns.catplot(x='sex', hue='survived', kind='count', data=Data)
```

```python
# Create a violin plot to show the distribution of survival (survived)
# across gender (sex), with different colors representing passenger class (pclass)
sns.catplot(x='sex', y='survived', hue='pclass', kind='violin', data=Data)
```

```python
# Create a boxen plot to visualize the distribution of survival (survived)
# across gender (sex), with different colors representing passenger class (pclass)
sns.catplot(x='sex', y='survived', hue='pclass', kind='boxen', data=Data)
```

```python
# Create a count plot showing the number of passengers by gender (sex)
# with survival status (survived) represented by different colors
sns.catplot(x="sex", hue="survived", kind="count", data=Data)
```

```python
# Create a bar plot to show the average survival rate (survived)
# for each gender (sex), with different colors representing passenger class (pclass)
sns.catplot(x="sex", y="survived", hue="pclass", kind="bar", data=Data)
```

```python
# Create a count plot to show the number of passengers by gender (sex)
# with different colors representing passenger class (pclass)
sns.catplot(x="sex", hue="pclass", kind="count", data=Data)
```

```python
# Check for missing (null) values in the dataset
# Returns True for missing values and False for non-missing values
Data.isnull()
```

```python
# Create count plots of passengers by embarkation point (embarked),
# showing survival status (survived) with separate plots for each passenger class (pclass)
sns.catplot(x="embarked", hue="survived", kind="count", col="pclass", data=Data)
```

```python
# Calculate the total number of male passengers who survived, grouped by passenger class (pclass)
Data[Data["sex"] == "male"].groupby("pclass")["survived"].sum()
```

```python
# Calculate the total number of female passengers who survived, grouped by passenger class (pclass)
Data[Data["sex"] == "female"].groupby("pclass")["survived"].sum()
```

```python
Data[Data["pclass"] == "male"].groupby("pclass")["survived"].sum()
```

```python
# Create a bar plot showing the average survival rate (survived)
# for each passenger class (pclass), with different colors representing gender (sex)
sns.catplot(
    x="pclass",
    y="survived",
    hue="sex",
    kind="bar",
    data=Data
)

# Display the plot
plt.show()
```

```python
# Count the number of female passengers who survived, grouped by travel class (class)
Data[
    (Data['sex'] == 'female') & (Data['survived'] == 1)
].groupby("class")["survived"].count()
```

```python
# Count the number of male passengers who survived, grouped by travel class (class)
Data[
    (Data['sex'] == 'male') & (Data['survived'] == 1)
].groupby("class")["survived"].count()
```

```python
# Count the number of female passengers who did not survive, grouped by travel class (class)
Data[
    (Data['sex'] == 'female') & (Data['survived'] == 0)
].groupby("class")["survived"].count()
```

```python
# Count the number of male passengers who did not survive, grouped by travel class (class)
Data[
    (Data['sex'] == 'male') & (Data['survived'] == 0)
].groupby("class")["survived"].count()
```

```python
# Create a heatmap to visualize missing (null) values in the dataset
# Darker/lighter colors indicate presence or absence of missing data
sns.heatmap(Data.isnull(), cmap="magma")
```
